```python
import os

import pandas as pd

engine = sqlalchemy.create_engine("postgresql://postgres:mysecretpassword@localhost:7432/hw4")

# Cutting df to 8000 rows because of local performance
RAW_PATH = "raw"
for raw_file in os.listdir(RAW_PATH):
    raw_df = pd.read_csv(os.path.join(RAW_PATH, raw_file), low_memory=True, nrows=8000)

    raw_df.to_csv(raw_file, index=False)

    table_name = raw_file.split(".")[0]
    raw_df.to_sql(table_name, engine, index=False, method="multi")

    print(raw_file.upper())
    print(raw_df.dtypes)
    print("=" * 20, end="\n\n\n")

```
